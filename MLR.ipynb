```python
import numpy as np
import pandas as pd
import seaborn as sns
# from sklearn.cross_validation import train_test_split
# it was renamed to --> model_selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_test_split
```

```python
dataset = pd.read_csv('startups.csv')
dataset.head()
```

```python
X = dataset.iloc[:,:-2]
y = dataset.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
type(y_predict)
```

```python
for i in range(len(y_predict)):
    print(np.array(y_test)[i],'\t', y_predict[i])
```

```python
sns.scatterplot(x = X_test['R&D Spend'], y = y_test)
sns.lineplot(x = X_test['R&D Spend'], y = y_predict, color='orange')
```

```python
sns.scatterplot(x = X_test['Marketing Spend'], y = y_test)
sns.lineplot(x = X_test['Marketing Spend'], y = y_predict, color='orange')
```

```python
sns.scatterplot(x = X_test['Administration'], y = y_test)
sns.lineplot(x = X_test['Administration'], y = y_predict, color='orange')
```

```python
model.score(X_test, y_test)
```
